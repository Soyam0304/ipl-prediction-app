```python
import pandas as pd
```

```python
dataset1=pd.read_csv("deliveries.csv")
dataset=pd.read_csv("matches.csv")
dataset1
```

```python
total_score=dataset1.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
total_score
```

```python
total_score = total_score[total_score['inning']==1]
```

```python
total_score
```

```python
match=dataset.merge(total_score[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
 match['team1'].unique()
```

```python
teams = ['Sunrisers Hyderabad', 'Mumbai Indians', 'Royal Challengers Bangalore',
       'Kolkata Knight Riders','Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals', 
       'Delhi Capitals']
```

```python
match['team1']=match['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match['team2']=match['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match['team1']=match['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match['team2']=match['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

```

```python
match = match[match['team1'].isin(teams)]
match = match[match['team2'].isin(teams)]
```

```python
match.head(2)
```

```python
match.shape
```

```python
#we are removing matches where duck wearth lewis is available 
# 1 = yes , 0 = no
match=match[match['dl_applied']==0]
```

```python
match.shape
```

```python
match=match[['match_id','city','winner','total_runs']]
```

```python
final=match.merge(dataset1,on='match_id')
```

```python
final.head()
```

```python
final = final[final['inning'] == 2]
```

```python
final.head()
```

```python
final.shape
```

```python
final.head(2)
```

```python
final['total_runs_y']
```

```python
final['total_runs_y'] = final['total_runs_y'].astype('int64')
```

```python
final['total_runs_y']
```

```python
final['current_score']=final.groupby('match_id')['total_runs_y'].cumsum()
```

```python
final['runs_left'] = final['total_runs_x']-final['current_score']
```

```python
final['balls left'] = 126 - (final['over']*6 + final['ball'])
```

```python
final.head(2)
```

```python
final['player_dismissed']=final["player_dismissed"].fillna(0)
```

```python
final.head(3)
```

```python
final['player_dismissed']=final['player_dismissed'].apply(lambda x : x if x == 0 else 1 )
final['player_dismissed'] = final['player_dismissed'].astype('int64')
```

```python
wickets = final.groupby('match_id')['player_dismissed'].cumsum().values
final['wicket'] = 10 - wickets
```

```python
final.head(3)
```

```python
final.tail()
```

```python
# current run rate = runs/overs
final['crr']=(final['current_score']*6)/(120 - final['balls left'])
```

```python
#rrr 
final['rrr'] = (final['runs_left']*6)/final['balls left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
final['result']=final.apply(result,axis=1)
```

```python
final1 =final[['batting_team','bowling_team','city','runs_left','balls left','wicket','total_runs_x','crr','rrr','result']]
```

```python
final1=final1.sample(final.shape[0])
```

```python
final1.dropna(inplace=True)
```

```python
final1.head(3)
```

```python
final1 = final1[final1['balls left']!=0]
```

```python
x = final1.iloc[:,:-1]
y = final1.iloc[:,-1]
```

```python
final1['city'].unique()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer(
    [('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])],
    remainder='passthrough'
)
```

```python
help(ColumnTransformer)
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.pipeline import Pipeline 
```

```python
pipe = Pipeline(steps=[('step1',trf),('step2',LogisticRegression(solver='liblinear'))])
```

```python
pipe.fit(x_train,y_train)
pipe.score(x_train,y_train)*100 , pipe.score(x_test,y_test)*100
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
accuracy_score(y_test,y_pred)*100
```

```python
def get_user_input():
    batting_team = input("Enter the batting team: ")
    bowling_team = input("Enter the bowling team: ")
    city = input("Enter the city: ")
    current_over = int(input("Enter the current over: "))
    
    overs_left = 20 - current_over 
    
    # Calculating derived inputs
    balls_left = overs_left * 6
    current_score = int(input("Enter the current score: "))
    target_score = int(input("Enter the target score: "))
    runs_left = target_score - current_score
    wic = int(input("Enter the number of wickets down: "))
    wickets_left = 10 - wic
    
    # Calculating crr and rrr
    crr = (current_score * 6) / (120 - balls_left)
    rrr = (runs_left * 6) / balls_left
    
    # Prepare the dataframe
    data = {
        'batting_team': [batting_team],
        'bowling_team': [bowling_team],
        'city': [city],
        'runs_left': [runs_left],
        'balls left': [balls_left],
        'wicket': [wickets_left],
        'total_runs_x': [target_score],
        'crr': [crr],
        'rrr': [rrr],
        'overs_left':[overs_left]
    }
    
    return pd.DataFrame(data)

def predict_match_outcome(pipe):
    
    new_match_df = get_user_input()
    
    # Predict probabilities using the trained pipeline
    result = pipe2.predict_proba(new_match_df)
    new_match_df['lose'] = np.round(result.T[0] * 100, 1)
    new_match_df['win'] = np.round(result.T[1] * 100, 1)
    #new = user
    #result = pipe.predict_prob(new)
    #new['loss'] = np.round(result.T[0] * 100,1)
    #new['win'] = np.round(result.T[1]*100,1)
    
   
    # Display the results
    print(new_match_df[['balls left','runs_left']])
    print(new_match_df[['overs_left','crr', 'rrr', 'win','lose']])
    
#predict_match_outcome(pipe)
 
```
